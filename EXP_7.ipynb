```python
#4. Fibonacci Series using while loop:
## method 1:
n= int(input("Enter number of terms:"))
a = 0
b = 1
i = 1
while i<=n:
  print(a, end=" ")
  c = a + b
  a = b
  b = c
  i += 1
```

```python
## method 2:
limit = int(input("Enter number of terms:"))
a = 0
b = 1
while a <= limit:
  print(a, end=" ")
  a,b = b, a + b
```

```python
#5. Reverse a number :
num = int(input("Enter a number:"))
rev = 0
while num>0:
  digit = num % 10
  rev = rev * 10 + digit
  num //= 10
print("Reversed number=", rev)
```

```python
#6. Check Palindrome number :
num = int(input("Enter a number:"))
temp = num
rev = 0
while num > 0:
  rev = rev * 10 + num % 10
  num //= 10
if temp == rev:
  print("Palindrome")
else:
  print("Not a Palindrome")
```

```python
#7. Generalized way to Check String Palindrome :
s = str(input("Enter a string:"))
i, j= 0, len(s)-1
is_palandrome = True
while i<j:
  if s[i] != s[j]:
    is_palandrome = False
    break
  i += 1
  j -= 1
if is_palandrome:
  print(f"{s} is a Palindrome")
else:
  print(f"{s} is not a Palindrome")
```

```python
#8. Check Palindrome using if-else: (works with both numbers and letters)
## numbers and lettes and stored as string
st = input("Enter a string:")
rev = st[::-1]
if st == rev:
  print("Palindrome")
else:
  print("Not a Palindrome")
```

```python
# Count digits in a number:
num = int(input("Enter a number:"))
count = 0
while num > 0:
  count += 1
  num //= 10
print("Number of digits=", count)
```

THE BREAK STATEMENT:
#
With the break statement we can stop the loop even if the while condition is True.

```python
#1. Exit the loop when i is 3:
i=1
while i<6:
  print(i)
  if i==3:
    break
  i+=1
```

```python
#2. Search an element in a list:
nums=[10,20, 30, 40, 50]
key = int(input("Enter element to search:"))
i = 0
while i< len(nums):
  if nums[i] == key:
    print("Element found in index", i + 1)
    break
  i += 1
else:
  print("Element not found")
```

The Continue Statement:

```python
#1. Print only odd numbers (1 to 10)
i = 0
while i < 10:
  i += 1
  if i % 2 == 0:
    continue
  print(i)
```
